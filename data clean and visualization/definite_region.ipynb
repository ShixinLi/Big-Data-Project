```python
import pandas as pd
```

```python
station_region=pd.read_csv('station_region.csv')
```

```python
definite_region_list=[360,454,483,495,498,509,3118,3135,3175,3232]
```

```python
df0=station_region[station_region['station_id']==definite_region_list[0]]
df1=station_region[station_region['station_id']==definite_region_list[1]]
df2=station_region[station_region['station_id']==definite_region_list[2]]
df3=station_region[station_region['station_id']==definite_region_list[3]]
df4=station_region[station_region['station_id']==definite_region_list[4]]
df5=station_region[station_region['station_id']==definite_region_list[5]]
df6=station_region[station_region['station_id']==definite_region_list[6]]
df7=station_region[station_region['station_id']==definite_region_list[7]]
df8=station_region[station_region['station_id']==definite_region_list[8]]
df9=station_region[station_region['station_id']==definite_region_list[9]]
```

```python
frame=[df0,df1,df2,df3,df4,df5,df6,df7,df8,df9]
result = pd.concat(frame)
```

```python
result.index=range(len(result))
```

```python
from geopy.distance import vincenty
station_list=list()
station_trip_list=list()
for i in range(len(station_region)):   
    min_mile=100000
    id_mile=-1
    for j in range(len(result)):
        mile_between=vincenty((station_region.iloc[i]['station_latitude'],station_region.iloc[i]['station_longitude']),(result.iloc[j]['station_latitude'],result.iloc[j]['station_longitude'])).miles
        if mile_between<min_mile:
            min_mile=mile_between
            id_mile=result.iloc[j]['station_id']
    station_list.append(id_mile)
```

```python
station_region
```

```python
station_region['neighbor_id']=station_list
```

```python
station_region_group_by_neighbor_id=station_region.groupby('neighbor_id')
```

```python
result_list=list()
for neighbor_id in definite_region_list:
    neighbor_df=station_region_group_by_neighbor_id.get_group(neighbor_id)
    result_list.append(sum(neighbor_df['number_of_trip']))
```

```python
result_list
```

```python
result['number_of_trip']=result_list
```

```python
result=result.sort(['number_of_trip'],ascending=False)
```

```python
result.to_csv('taxi_trip_by_station_id.csv')
```
