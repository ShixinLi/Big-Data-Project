```python
from geopy.geocoders import Nominatim
from geopy.distance import vincenty
import random
import pandas as pd
```

```python
citibike_tripdata=pd.read_csv("data/201512-citibike-tripdata.csv")
station_id_list=list(citibike_tripdata['start station id'].unique())
start_station_latitude_list=list(citibike_tripdata['start station latitude'].unique())
start_station_longitude_list=list(citibike_tripdata['start station longitude'].unique())
```

```python
for count in range(3,10):
    yellow_tripdata=pd.read_csv("data/yellow_tripdata_2015-0"+str(count)+".csv")
    yellow_tripdata=yellow_tripdata[yellow_tripdata['pickup_latitude']!=0]
    rows = random.sample(yellow_tripdata.index, 20000)
    yellow_tripdata=yellow_tripdata.ix[rows]
    distance_list=list();
    citibike_station_id_list=list()
    citibike_station_latitude_list=list()
    citibike_station_longitude_list=list()
    for i in range(len(yellow_tripdata)):
        min_mile=1000000
        station_id=-1
        for j in range(len(station_id_list)):
            mile_between=vincenty((start_station_latitude_list[j],start_station_longitude_list[j]),(yellow_tripdata.iloc[i,:]['dropoff_latitude'],yellow_tripdata.iloc[i,:]['dropoff_longitude'])).miles
            if mile_between<min_mile:
                station_id=j
                min_mile=mile_between
        print str(count)+"="+str(i)
        distance_list.append(min_mile)  
        citibike_station_id_list.append(station_id)
        citibike_station_latitude_list.append(start_station_latitude_list[station_id])
        citibike_station_longitude_list.append(start_station_longitude_list[station_id])
    yellow_tripdata.index=range(20000)
    yellow_tripdata['station_id']=pd.Series(citibike_station_id_list,dtype=int)
    yellow_tripdata['station_latitude']=pd.Series(citibike_station_latitude_list,dtype=float)
    yellow_tripdata['station_longitude']=pd.Series(citibike_station_longitude_list,dtype=float)
    yellow_tripdata.to_csv('yellow_tripdata_0'+str(count)+'_20000.csv')
```
